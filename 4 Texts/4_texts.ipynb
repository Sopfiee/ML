```python
stop_w=list(punctuation)  + nltk.corpus.stopwords.words('russian')
stop_w
```

```python
from pymorphy2 import MorphAnalyzer
```

```python
morph=MorphAnalyzer()
```

```python
def my_tokenizer(text):
    tokens=[w for w in nltk.word_tokenize(text.lower())
           if (w not in stop_w) 
           and not re.match('.*[.\'`*/a-z0-9]', w)]
    return [morph.parse(w)[0].normal_form for w in tokens]
```

```python
def my_new_tokenizer(text):
    tokens=[w for w in nltk.sent_tokenize(text.lower())]#предложения
    tokens1=[]
    for i in tokens:
        t=[w for w in nltk.word_tokenize(i)
          if (w not in stop_w) and not re.match('.*[.\'`*/a-z0-9]', w)] # слова в предложении
        tokens1+=[morph.parse(w)[0].normal_form for w in t]
    return tokens1
```

```python
def text_to_dataset(url, label):
    #получаем тескт произведения
    html = urlopen(url).read().decode('cp1251').encode('utf-8')
    soup = BeautifulSoup(html, features='html.parser')
    author_text = soup.body.get_text()
    #в датасете будет колонка 'predl' содержащая предложения произведений
    predlojeniya = [w for w in nltk.sent_tokenize(author_text.lower())]
    #удалим первые и последние 10 предложений, так как там может быть указано имя автора, дата написания и т. д.
    del predlojeniya[0:10] 
    for i in range(10):
        del predlojeniya[len(predlojeniya)-1]
    
    #также в датасете будет колонка label, показывающая, относится ли предложение к Куприну или к кому-то другому
    df = pd.DataFrame(predlojeniya,columns=['predl'])
    df['label'] = [label] * len(df)
    return df
```

```python
kuprin=text_to_dataset('http://az.lib.ru/k/kuprin_a_i/text_0130.shtml','kuprin') #датасет - Куприн "Олеся"
```

```python
kuprin
```

```python
dostoevsky=text_to_dataset('http://az.lib.ru/d/dostoewskij_f_m/text_0060.shtml','other')#датасет достоевский
tolstoy=text_to_dataset('http://az.lib.ru/t/tolstoj_lew_nikolaewich/text_0040.shtml','other')#датасет толстой
```

```python
authors = kuprin.append([dostoevsky, tolstoy])
```

```python
authors
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test =train_test_split(authors.predl, authors.label)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer(ngram_range=(1,2), tokenizer=my_new_tokenizer)
```

```python
x_t = tfidf.fit_transform(authors.predl)
```

```python
tfidf.vocabulary_
```

```python
x_train_t=tfidf.transform(x_train)
x_test_t=tfidf.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
logreg=LogisticRegression()
```

```python
logreg.fit(x_train_t, y_train)
```

```python
print(classification_report(y_test, logreg.predict(x_test_t)))
# при полном токенайзере с stop_w, re.match() и normal_form при ngram =(1,1) - 0.94, 0.9352038554915667
# при токенайзере с re.match() и normal_form при ngram =(1,1) - 0.95, 0.9483276758340381
# при полном токенайзере, ngram=(1,2) - 0.93, 0.9311721255409647
```

```python
from sklearn.model_selection import cross_val_score
np.mean(cross_val_score(logreg,x_t, authors.label, cv=4))
```

```python
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
```

```python
dtc=DecisionTreeClassifier(criterion='gini', max_depth=100, min_samples_split=10)
# при полном токенайзере с stop_w, re.match() и normal_form при ngram =(1,1)
# max-depth = 100 - 0.94, 0.9362323099442604,
#              50 - 0.95, 0.9403874425127692
# при полном токенайзере, ngram=(1,2)
# max-depth = 100 - 0.94, 0.9372608050211012
#              50 - 0.94, 0.941950754487912
```

```python
dtc.fit(x_train_t, y_train)
```

```python
print(classification_report(y_test, dtc.predict(x_test_t)))
```

```python
np.mean(cross_val_score(dtc, x_t, authors.label, cv=4))
```
